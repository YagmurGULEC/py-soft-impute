```python
import numpy as np
import soft_impute 
import matplotlib.pyplot as plt


def generate_data(size_x=50,size_y=50,no_nans=20):
    
    X=np.random.randn(size_x,size_y)
    
    for i in range(no_nans):
        X[np.random.randint(0,X.shape[0]),np.random.randint(0,X.shape[1])]=np.nan
    return X
X=generate_data()
```

```python
clf=soft_impute.Impute(maxit=400)

clf.fit(X)
```
